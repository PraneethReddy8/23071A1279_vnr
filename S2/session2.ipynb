```python
df = pd.read_csv(r"C:\Users\Reddy\Downloads\data.csv")  
df.head()
```

```python

df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df = df.drop(columns=['Deaths / 100 Recovered'])
```

```python
df = df[df['Confirmed'] >= 100]
```

```python

df.to_csv("cleaned_data.csv", index=False)

df.to_excel("cleaned_data.xlsx", index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convert "Unknown" back to NaN for numerical columns
df.replace("Unknown", np.nan, inplace=True)

# Convert 'Deaths' and 'Confirmed' to numeric type
df['Confirmed'] = pd.to_numeric(df['Confirmed'], errors='coerce')
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')

# Fill remaining NaN values with column mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Now create the scatter plot
plt.figure(figsize=(8,5))
plt.scatter(df['Confirmed'], df['Deaths'], color='red', alpha=0.5)
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.title('Confirmed Cases vs Deaths')
plt.show()
```

```python
import seaborn as sns

# Box plot: Active cases per WHO Region
plt.figure(figsize=(10,6))
sns.boxplot(x='WHO Region', y='Active', data=df)
plt.xticks(rotation=45)
plt.title('Distribution of Active Cases Across WHO Regions')
plt.show()
```

```python
# Bar chart: New cases by WHO Region
df.groupby('WHO Region')['New cases'].sum().plot(kind='bar', figsize=(10,5), color='blue')

plt.xlabel('WHO Region')
plt.ylabel('Total New Cases')
plt.title('New Cases Across WHO Regions')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sort by confirmed cases (optional, for better visualization)
df_sorted = df.sort_values(by="Confirmed last week")

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(df_sorted['Country/Region'], df_sorted['Confirmed'], label="Confirmed Cases (Current)", marker='o', linestyle='-')
plt.plot(df_sorted['Country/Region'], df_sorted['Confirmed last week'], label="Confirmed Cases (Last Week)", marker='s', linestyle='--')

# Rotate x-axis labels for readability
plt.xticks(rotation=90, fontsize=10)
plt.yticks(fontsize=10)

# Add labels and title
plt.xlabel("Country/Region", fontsize=12)
plt.ylabel("Number of Confirmed Cases", fontsize=12)
plt.title("Confirmed Cases vs Last Week", fontsize=14)
plt.legend()

# Show the plot
plt.show()
```
